```python
class Guardia:
    def __init__(self, nombre, telefono, edad, honestidad, comunicacion, disciplina):
        # Atributos de identidad
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

        # Atributos de perfil (referentes a la clase)
        self.honestidad = honestidad
        self.comunicacion = comunicacion
        self.disciplina = disciplina

    # Funciones (Métodos)
    def vigila(self):
        print(f"[{self.nombre}]: Está realizando un recorrido de vigilancia. Nivel de disciplina: {self.disciplina}/10.")

    def cuida(self, lugar):
        print(f"[{self.nombre}]: Se encuentra custodiando el acceso a {lugar}. Reporte: Sin novedades.")

    def reportes(self, mensaje):
        print(f"[{self.nombre}] CENTRAL: '{mensaje}' (Enviado vía radio).")

# --- Creación de Objetos (Instancias) ---

# Objeto 1: Asignado a Radio 1
radio_1 = Guardia("Oficial Pérez", "555-0123", 35, 10, 9, 10)

# Objeto 2: Asignado a Chaleco 1
chaleco_1 = Guardia("Oficial Gómez", "555-0987", 28, 9, 8, 9)

# --- Ejecución de ejemplos ---

print("--- SIMULACIÓN DE SEGURIDAD ---")
radio_1.vigila()
radio_1.reportes("Todo despejado en el sector norte.")

print("-" * 30)

chaleco_1.cuida("Almacén Principal")
chaleco_1.reportes("Se detectó un candado flojo, procedo a asegurar.")

print("Programa terminado")
print("programa realizado por santiago Pacheco NL = 43")
```
